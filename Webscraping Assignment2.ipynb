# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. 



```python
#First get the webpage https://www.shine.com/

driver = webdriver.Chrome()
driver.get('https://www.shine.com/')
```

```python
#Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.

designation = driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys('Data Analyst')
location = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')

#Then click the searchbutton.

search = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()
```

```python
job_title = []
job_location = []
company_name =[]
experience_required = []
```

```python
#Scraping data for first 10 job results and storing them inside respective lists

title_tags = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
job_title = [tag.text for tag in title_tags[0:10]]


loc1= []
loc = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc[0:10]:
    title = i.text
    loc1.append(title)

substrings = ['\n+14','\n+9','\n+14']

for element in loc1:
    for substring in substrings:
        element = element.replace(substring, '')
    job_location.append(element)

name_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
company_name = [i.text for i in name_tags[0:10]]

exp_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
experience_required = [i.text for i in exp_tags[0:10]]


```

```python
#creating a dataframe for the same

df = pd.DataFrame({'Job Title': job_title, 'Job Location': job_location, 'Company Name':company_name, 'Experience Required':experience_required},index = range(1,11))
df
```
